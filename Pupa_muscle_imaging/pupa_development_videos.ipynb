## Script to makevideos out of Leandre's pngs of pupal development

```python
import os
import glob
import cv2
```

```python
# find data
main_path = '/mnt/upramdya_data/LD/Data/PupAgeNet'

# create output folder
output_folder = os.path.join(main_path, 'Videos')
os.makedirs(output_folder, exist_ok=True)
```

```python
# Function to make video (mp4) from all images in batch folder (png)
def create_video_from_pngs(image_folder, output_file, fps=10.0):
    """
    Creates an MP4 video from a sequence of PNG files in a folder.

    Args:
        image_folder (str): Path to the folder containing the PNG images.
        output_file (str): Path and filename for the output video (e.g., 'output.mp4').
        fps (float): Frames per second for the output video.
    """
    print(f"Starting video generation for folder: {image_folder}")

    # 1. Gather all PNG files and sort them numerically/lexicographically
    # This ensures the frames are processed in the correct order (e.g., 1.png, 2.png, 10.png)
    image_paths = sorted(glob.glob(os.path.join(image_folder, '*.png')))

    if not image_paths:
        print("Error: No PNG files found in the specified folder.")
        return

    # 2. Read the first image to get the frame size
    first_image = cv2.imread(image_paths[0])
    if first_image is None:
        print(f"Error: Could not read the first image file: {image_paths[0]}")
        return

    height, width, layers = first_image.shape
    frame_size = (width, height)
    print(f"Determined frame size: {frame_size}, Total frames to process: {len(image_paths)}")

    # 3. Define the codec and create VideoWriter object
    # The 'mp4v' codec (fourcc) is generally reliable for creating MP4 files.
    # FourCC is a 4-character code used to specify the video codec.
    fourcc = cv2.VideoWriter_fourcc(*'mp4v') 
    
    # Initialize the video writer
    out = cv2.VideoWriter(output_file, fourcc, fps, frame_size)

    if not out.isOpened():
        print(f"Error: VideoWriter failed to open for file: {output_file}")
        return

    # 4. Loop through all images and write them to the video file
    for i, image_path in enumerate(image_paths):
        # Read the image
        frame = cv2.imread(image_path)
        
        # Check if the image was read successfully
        if frame is None:
            print(f"Warning: Skipped unreadable file: {image_path}")
            continue

        # Check if the frame size matches the initialized video size
        if frame.shape[1] != width or frame.shape[0] != height:
            print(f"Warning: Frame size mismatch for {image_path}. Resizing...")
            frame = cv2.resize(frame, frame_size)

        # Write the frame to the output video
        out.write(frame)

        # Print progress every 100 frames
        if (i + 1) % 100 == 0:
            print(f"Processed {i + 1}/{len(image_paths)} frames...")

    # 5. Release the VideoWriter object
    out.release()
    print("---------------------------------------")
    print(f"Successfully created video: {output_file}")
    print(f"Total frames processed: {len(image_paths)}")
    print("---------------------------------------")
```

```python
for n in range(7):
    print(f'Processing Batch {n+1}...')
    batch_path = os.path.join(main_path, f'Batch{n+1}')
    output_file = os.path.join(output_folder, f'pupa_development_video_batch{n+1}.mp4')
    create_video_from_pngs(batch_path, output_file, fps=10.0)
```
